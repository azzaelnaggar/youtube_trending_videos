```python
import os
import pandas as pd
import numpy as np
```

```python
dir_name = r'datacsv/'
list_of_files = sorted( os.listdir(dir_name),
                        key = lambda x: os.path.getmtime(os.path.join(dir_name, x))
                        )
```

```python
import requests
import sys
channel_id = 'UCXuqSBlHAE6Xw-yeJA0Tunw'
#api_key="AIzaSyAWgN_2JVet9ZvsZ5ELw_ncLK5bgUvBL2k"
def add_channel_info(channel_id,api_key):
    request_url = f"https://youtube.googleapis.com/youtube/v3/channels?part=snippet%2Cstatistics%2CbrandingSettings%2CtopicDetails&id={channel_id}&key={api_key}"
        #print(request_url)
    request = requests.get(request_url)
    if request.status_code != 200:
        print(f"errrooorrrrrrrrr_-_in country {channel_id}")
        sys.exit()
    Items ={} 
    if 'items' in request.json():
        Items = request.json()['items']
    else:
        return []

    df_dict_list =[]
    diict={}
    for item in Items:
        #print(type(item['id']))
        snip = item['snippet']
        stat= item['statistics']
        
            
        diict.update({'publishedAt':snip['publishedAt'], "title":snip["title"],"thumbnails":snip["thumbnails"]['medium']['url'],
        "description":snip["description"]})
        if 'country' in item['snippet']:
            diict .update({'country': snip['country']})
        if "channelId" in snip:
            diict .update({'channelId': snip['channelId']})
        else:
            diict .update({'channelId': item['id']})
        if 'image' in item['brandingSettings']:
            diict.update({"cover_image":item['brandingSettings']['image']['bannerExternalUrl']})
        if "viewCount" in stat:
            diict.update({"viewCount":stat["viewCount"]})
        if "subscriberCount" in stat:
            diict.update({"subscriberCount":stat["subscriberCount"]})
        if "videoCount" in stat:
            diict.update({"videoCount":stat["videoCount"]})
        if 'topicDetails' in item:
            if 'topicIds' in item['topicDetails']:
                diict .update({"topicIds":item['topicDetails']['topicIds'],})
        df_dict_list.append(diict)
    return df_dict_list

```
