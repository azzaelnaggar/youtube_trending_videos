```python
import pandas as pd 
```

```python

#csv files name in data folder
files = ['US_youtube_trending_data.csv','GB_youtube_trending_data.csv',
'CA_youtube_trending_data.csv','RU_youtube_trending_data.csv']
#country name for each csv file
country = ['usa', 'birtain', 'canda', 'russia']
channels = []
dataframes = []

for i in range(len(files)):
    #import data csv file     
    df = pd.read_csv(f'../data/{files[i]}')
    #get unique channel id
    channels.extend(list(df['channelId'].unique()))
    #drop not needed columns and columns not supported any more by youtube
    df.drop(columns=['dislikes','comments_disabled','ratings_disabled'],inplace=True)
    #add country name to dataframe for join all files together
    df["regions"]=country[i]

    dataframes.append(df)
    #save file after edit 
    df.to_csv(f'../data/{files[i]}',index=False)

#concat all dataframes and save to one csv file
final_df = pd.concat(dataframes)
final_df.to_csv(f'../data/all_world.csv',index=False)

#convert channel id to dataframe and save it to csv file
channels_id = pd.DataFrame(channels, columns=['channelId'])
channels_id.to_csv(f'../data/channels_id.csv',index=False)
```
